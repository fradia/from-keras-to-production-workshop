# Images & Matplotlib

- Create a list of file names in the validation directory and
- print randomly 16 images:

```python
train = '/keras2production/fruits/Training/'
```

```python
images = !find $train -type f -name '*.jpg'
```

```python
num_images = 16
images_sample = random.sample(images, num_images)

fig = plt.figure(figsize=(12,8))
for item in images_sample:
    fig.add_subplot(4,4,images_sample.index(item)+1)
    im = plt.imread(item)
    plt.imshow(im)
    plt.axis('off')
```
